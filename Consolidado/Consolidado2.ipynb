```python
import pandas as pd
```

```python
df = pd.read_excel(r"D:\github\DATA-EMERGENCY\Consolidado\Consolidado.xlsx")
df
```

```python
df["Urgencia"] = df["Urgencia"].apply(lambda x : x.replace("-\xa0","").replace("- ",""))
```

```python
diccionario = {}
for i in df["Urgencia"].unique():
    diccionario[i] = i    
```

```python
diccionario
```

```python
diccionario["IRA Alta (J00-J06)"] = "SISTEMA RESPIRATORIO"
diccionario["Influenza (J09-J11)"] = "SISTEMA RESPIRATORIO"
diccionario["Bronquitis/bronquiolitis aguda (J20-J21)"] = "SISTEMA RESPIRATORIO"
diccionario["Crisis obstructiva bronquial (J40-J46)"] = "SISTEMA RESPIRATORIO"
diccionario["Otra causa respiratoria (J22, J30-J39, J47, J60-J98)"] = "SISTEMA RESPIRATORIO"

diccionario["COVID 19 Sospechoso (U072)"] = "SOSPECHA COVID 19"
diccionario["COVID 19 Confirmado (U071)"] = "CONFIRMADO COVID 19"

diccionario["Infarto agudo miocardio"] = "SISTEMA CIRCULATORIO"
diccionario["Accidente vascular encefálico"] = "SISTEMA CIRCULATORIO"
diccionario["Crisis hipertensiva"] = "SISTEMA CIRCULATORIO"
diccionario["Arritmia grave"] = "SISTEMA CIRCULATORIO"
diccionario["Otras causas circulatorias"] = "SISTEMA CIRCULATORIO"


diccionario['CAUSAS SISTEMA CIRCULATORIO'] = "SISTEMA CIRCULATORIO"
diccionario['CAUSAS SISTEMA RESPIRATORIO'] = "SISTEMA RESPIRATORIO"
diccionario['SOSPECHA CORONAVIRUS (U072)'] = "SOSPECHA COVID 19"
diccionario['CORONAVIRUS (U071)']          = "CONFIRMADO COVID 19"


diccionario['LAS DEMÁS CAUSAS']          = "TOTAL DEMÁS CAUSAS"

diccionario


```

```python
df["Causas"] = df["Urgencia"].apply(lambda x: diccionario[x])
```

```python
df.to_excel(r"D:\github\DATA-EMERGENCY\Consolidado\Consolidado.xlsx", index=False)
```
