```python
import os
import pandas as pd
```

```python

dirlist = os.listdir("carga urgencias")
files = []
for i in dirlist:
    if(".csv" in i):
        files.append(i)
```

```python
tipo = []
for i in range(17):
    tipo.append("URGENCIA")
for i in range(7):
    tipo.append("HOSPITALIZACIONES")
```

```python
salida = []
for i in files:
    #df = pd.read_csv(f"carga urgencias/{files[0]}")
    df = pd.read_csv(f"carga urgencias/{i}")
    #Suma
    #TOTAL CAUSAS SISTEMA RESPIRATORIO
    #TOTAL CAUSAS SISTEMA CIRCULATORIO
    #TOTAL TRAUMATISMOS Y ENVENENAMIENTO
    #SECCIÓN 2 TOTAL DE HOSPITALIZACIONES
    #SECCIÓN 1 TOTAL ATENCIONES DE URGENCIA
    df2 = df.drop([0,1, 2,11,17,22],axis=0)
    df2["Tipo"] = tipo
    #df
    #print(len(df))
    salida.append(df2)
```

```python
dfFinal = pd.concat(salida)
print(len(dfFinal))
```

```python
def ToInt(num):
    try:
        return int(num)
    except:
        return 0

del dfFinal["Total"]
for i in dfFinal.columns[1:53]:
    dfFinal[i] = dfFinal[i].apply(ToInt)
```

```python
dfFinal.to_excel("Consolidado.xlsx", index=False)
```
