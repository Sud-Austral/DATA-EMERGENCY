```python
# El término Gecko significa un motor de navegador web integrado en el navegador Mozilla Firefox. 
# El controlador Gecko actúa como un proxy entre los clientes habilitados para el controlador web 
#(Eclipse, Netbeans, etc.) y el navegador Mozilla Firefox. En resumen, el controlador Gecko actúa 
# como un enlace entre las pruebas del controlador web Selenium y el navegador Mozilla Firefox.

# Antes de Selenium 3, el navegador Mozilla Firefox era el navegador predeterminado para Selenium. 
# Después de Selenium 3, los probadores deben inicializar el script para usar Firefox usando GeckoDriver 
# explícitamente. Selenium usa el protocolo W3C Webdriver para enviar solicitudes a GeckoDriver, que las 
# traduce a un protocolo llamado Marionette. Firefox comprenderá los comandos transmitidos en forma de 
# protocolo Marionette y los ejecutará.

# https://www.guru99.com/gecko-marionette-driver-selenium.html

from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.firefox.firefox_profile import FirefoxProfile
from selenium.webdriver.firefox.options import Options
import http.client, urllib.request, urllib.parse, urllib.error, base64
import shutil
import datetime
from selenium.webdriver.support.ui import Select
import time
import pandas as pd
```

```python
driver = getDriver()

element = driver.find_elements_by_tag_name("select")

anios = element[0]
regiones = element[2]
tipoEstablecimiento = element[3]

# se llenan los arreglos valor con los valores concretos de cada seleccion:
valorAnio = anios.find_elements_by_tag_name("option")
valorRegiones = regiones.find_elements_by_tag_name("option")
valorTipoEstablecimiento = tipoEstablecimiento.find_elements_by_tag_name("option")

# Select se ha importado desde Selenium

selectAnios = Select(anios)
selectregiones = Select(regiones)
selectTipoEstablecimiento = Select(tipoEstablecimiento)

for i in valorAnio:
    
    selectAnios.deselect_all()
    
    # select_by_visible_text selecciona todas las opciones que muestren texto que coincidan con el argumento.
    
    selectAnios.select_by_visible_text(i.text)
    
    for j in valorRegiones:
        
        selectregiones.deselect_all()
        selectregiones.select_by_visible_text(j.text)
        
        for k in valorTipoEstablecimiento:
            selectTipoEstablecimiento.deselect_all()
            selectTipoEstablecimiento.select_by_visible_text(k.text)
            establecimiento = driver.find_elements_by_tag_name("select")[4]
            selectEstablecimiento = Select(establecimiento)
            valorEstablecimiento = establecimiento.find_elements_by_tag_name("option")
            
            for l in valorEstablecimiento:
                selectEstablecimiento.deselect_all()
                selectEstablecimiento.select_by_visible_text(l.text)

                try:
                    anio = i.text
                    region = j.text
                    tipo_Establecimiento = k.text
                    nombre_establecimiento = l.text
                    selectEstablecimiento.deselect_all()
                    selectEstablecimiento.select_by_visible_text(l.text)
                    nombreArchivo = f"{i.text}_{j.text}_{k.text}_{l.text}.csv"
                    dr = getDriver()
                    #print(getTable(dr,i,j,k,l)[0].get_attribute('innerHTML'))
                    
                    # se llama a la tabla cargada y se la lleva a un csv
                    
                    saveCSV(getTable(dr,i,j,k,l),nombreArchivo)
                    
                    dr.close()
                except:
                    anio = i.text
                    region = j.text
                    tipo_Establecimiento = k.text
                    nombre_establecimiento = l.text
                    nombreArchivo = f"{i.text}_{j.text}_{k.text}_{l.text}.csv"
                    print(nombreArchivo)
```
